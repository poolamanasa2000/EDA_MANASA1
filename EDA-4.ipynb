```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
visa_df=pd.read_csv('Visadataset.csv')
```

```python
case_status=visa_df['case_status'].unique()
count=[]
for i in visa_df['case_status'].unique():
    con=visa_df['case_status']==i
    count.append(len(visa_df[con]))

cols=['case_status','Number of Application']
pd.DataFrame(zip(case_status,count),columns=cols)

    
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming visa_df is your DataFrame and 'case_status' is the column you want to count
cdf = visa_df['case_status'].value_counts()
keys = cdf.index  # Use index to get the labels
values = cdf.values  # Use values to get the counts

# Create a new DataFrame with labels and counts
cols = ['Labels', 'Count']
df = pd.DataFrame(list(zip(keys, values)), columns=cols)

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.bar(df['Labels'], df['Count'], color='blue')
plt.xlabel('Labels')
plt.ylabel('Count')
plt.title('Count of Each Case Status')
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

**Bar chart using Searborn**

- We can plot bar charts using with many packages

- one packages is matplotlib

- Another package is seaborn

- In matplotlib the method is **plt.bar**

- In seaborn the method is **countplot**

- in matplotlib,

  - we will take raw data

  - we will create our frequency table i.e.dataframe

  - we will pass the column1,column2 and dataframe name

- in seaborn count plot

  - it takes direct raw value

  - we will provide direct column name

  - we will providevmain dataframe name
